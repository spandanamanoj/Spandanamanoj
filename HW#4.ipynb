# Write a function which can return a Factorial of any numbers as INT, given in the argument.

```python
def fact(n):
    n = int(input("Enter the number which you want the factorial of it's:"))
    a = 1
    for i in range(n,0,-1):
        a = a * (i)
    print("The factorial of a given no.", n, " is :", a)
    return(a)
```
